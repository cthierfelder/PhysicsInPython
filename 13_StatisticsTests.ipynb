```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.stats import chi2
from scipy.stats import gamma
from scipy.stats import t
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def genGaussNumbers(mu, sigma, N):
    return np.random.normal(loc=mu, scale=sigma, size=N)

def sumOfSquaredGaussNumbers(X):
    return np.sum(np.square(X))
```

```python
N=10000
mu=0.0
sigma=3.0
X = genGaussNumbers(mu, sigma,N)
```

```python
n, bins, patches = plt.hist(X, 50, density=True, facecolor='green', alpha=0.75)
l = plt.plot(bins, norm.pdf(bins,mu,sigma), 'r--', linewidth=1)
plt.xlabel('X')
plt.ylabel('Probability')
#plt.axis([-(mu+3*sigma), mu+3*sigma, 0, max(1.2*y)])
plt.grid(True)

plt.show()
```

## t-test

test if mean mu of a iid Gauss sample is mu0 - sqrt(n)*(xbar-mu0)/s is Student-t distributed

```python
M   = 10
mu    = 0.0
sigma = 2.0
mu0   = 0.0
T   = np.array([])
for y in [genGaussNumbers(mu, sigma, M) for i in range(10000)]:
    n    = len(y)
    xbar = np.mean(y)
    s    = np.std(y)*np.sqrt(n)/np.sqrt(n-1)
    tmp  = np.sqrt(n)*(xbar-mu0)/s
    T = np.append(T,tmp)
n, bins, patches = plt.hist(T, 50, density=True, facecolor='green', alpha=0.75)
l = plt.plot(bins, t.pdf(bins,M), 'r--', linewidth=1)
plt.xlabel('sum X^2')
plt.ylabel('Probability')
#plt.axis([0, 100, 0, 0.05])
plt.grid(True)

plt.show()
```
